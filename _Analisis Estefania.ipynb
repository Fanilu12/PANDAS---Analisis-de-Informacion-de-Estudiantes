## Correlation Analysis
### Matriz de Correlación de las Métricas de Rendimiento Estudiantil
En este paso, calculamos y visualizamos la matriz de correlación para analizar las relaciones entre diferentes variables del conjunto de datos.

### Detalles del análisis:
1. **Cálculo de la matriz de correlación**:

### `sns.boxplot`
`sns.boxplot` es una función de Seaborn utilizada para visualizar la distribución de datos y resumir estadísticamente las características de una variable en diferentes categorías.
#### Elementos clave del gráfico:
1. **Caja**:
   - Representa el rango intercuartílico (IQR), que es la distancia entre el primer cuartil (Q1, 25%) y el tercer cuartil (Q3, 75%).
2. **Línea en la caja**:
   - Indica la mediana (50%) de los datos.
3. **Bigotes (whiskers)**:
   - Muestran la extensión de los datos, excluyendo los valores atípicos.
4. **Puntos fuera de los bigotes**:
   - Representan valores atípicos (outliers), es decir, datos que están significativamente alejados de la distribución principal.

#### Parámetros útiles:
- **`x` y `y`**: Columnas del DataFrame que se representan en los ejes.
- **`hue`**: Agrupa los datos por una tercera variable categórica.
- **`palette`**: Define los colores utilizados en el gráfico.

### Ejemplo:
sns.boxplot(x='gender', y='math_score', data=data, palette="Blues", hue='gender')

### `sns.heatmap`
`sns.heatmap` es una función de Seaborn que representa datos matriciales (como una matriz de correlación) en forma de mapa de calor.

#### Elementos clave del gráfico:
1. **Celdas coloreadas**:
   - Representan valores numéricos mediante una escala de colores (`cmap`).
2. **Anotaciones (opcional)**:
   - Muestra los valores numéricos directamente en las celdas al habilitar `annot=True`.
3. **Escala de colores**:
   - Proporciona un gradiente visual que facilita la interpretación de los valores.

#### Parámetros útiles:
- **`data`**: Matriz o DataFrame a visualizar.
- **`annot`**: Habilita la visualización de los valores en cada celda.
- **`cmap`**: Define la paleta de colores, como `"coolwarm"`, `"bwr"`, o `"viridis"`.
- **`vmin` y `vmax`**: Ajustan el rango mínimo y máximo de valores para la escala de colores.
- **`linewidths`**: Controla el grosor de las líneas entre las celdas.

#### Ejemplo de uso:
Visualizamos una matriz de correlación utilizando `sns.heatmap` para analizar relaciones entre variables.

#### Calcular la matriz de correlación
correlation_matrix = data[['math_score', 'reading_score', 'writing_score', 'average_score']].corr()

#### Crear un heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1, linewidths=0.5)
